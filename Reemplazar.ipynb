```python
import random # Del 0. al 1. (https://docs.python.org/3/library/random.html)
import pandas as pd # Pandas se usa para trabajar con archivos de datos
import re # Regex para agarrar los espacios en blanco

from IPython.display import display, HTML # Para mostrar la tabla, no es necesario
```

```python
# CONDICIONES
# Asumiendo que se llama: np561.data
# Se usa regex para seleccionar los espacios en blanco en 'sep' para separar las columnas
# El archivo de datos no tiene encabezado
df = pd.read_csv('np561.data', sep='\s+', header=None)
```

```python
# Ahora, podemos llamar como un conjunto estructurado de datos
display(df)
```

```python
# Si queremos escoger solo la segunda columna:
display(df[1])
```

```python
# Iniciamos comparando los casos donde el valor de la segunda columna es exactamente '1'
# Sumamos la cantidad de veces donde esta comparación es 'True'
cantidad_de_unos = (df[1] == 1).sum()
print(f"Número de '1's en la segunda columna: {cantidad_de_unos}")
```

```python
# Asignamos un índice a cada valor de 1 en cada columna
# Esta lista se usará después para seleccionar una cantidad específica
# de índices random
índice_valor = df[df[1] == 1].index.tolist()
display(índice_valor)
```

```python
# Seed reproducible
random.seed(42)

# Usamos random.sample (https://docs.python.org/3/library/random.html#random.sample)
# para especificar la cantidad de índices que vamos a reemplazar
indices_a_reemplazar = random.sample(índice_valor, 56)

# Reemplazamos aquellos índices con 2
df.loc[indices_a_reemplazar, 1] = 2
```

```python
# Guardamos el archivo en txt
df.to_csv('np561_reemplazado.data', sep=' ', header=None, index=False)
```

```python
# Para asegurarnos de que todo va bien, veamos esta tabla más de cerca
display(df)
```

```python
# Contemos la cantidad de veces que el número '2' está siendo usado
# en la segunda columna
cantidad_de_unos = (df[1] == 2).sum()
print(f"Número de '2's en la segunda columna: {cantidad_de_unos}")
```
